```python
import os
import magic
from PIL import Image
```

---

```python
# get files in directory
def get_files(src_dir):
    f = []
    for (dirpath, dirnames, filenames) in os.walk(src_dir):
        f.extend(filenames)
        break
    return f
```

---

```python
INPUT_DIRS = [
    '/datasets/kaggle/painters/train',
    '/datasets/kaggle/painters/test'
]
```

```python
i = 0

for d in INPUT_DIRS:
    files = get_files(d)  

    for f in files:
        i = i + 1
        if i % 5000 == 0: print i
        
        cf = d + '/' + f   
        mime = magic.from_file(cf, mime=True)
        
        if 'image/jpeg' != mime:
            img = Image.open(cf)
            if img.mode != 'RGB': img = img.convert('RGB')
            cf_correct = cf + '.correct.jpg'
            img.save(cf_correct)
            print '* ', mime, cf, '->', cf_correct
```
